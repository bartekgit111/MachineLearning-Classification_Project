```python
import joblib

from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


from preprocessors import preprocessor_rf, preprocessor_et, preprocessor_gbc, preprocessor_xgb, preprocessor_lgbm
from model_tuner import model_tuner
from param_grids import param_grids
```

```python
models = {}
```

----
## Komitety
----

### 1. Baggign (glm, svm, knn, tree)

```python
basic_models = joblib.load("all_basic_grids.pkl")
glm_en_select = basic_models["glm_en_select"][0]
svm_PCA = basic_models["svm_PCA"][0]
knn_select = basic_models["knn_select"][0]
tree_select = basic_models["tree_select"][0]
```

```python
models["glm_en_bagging"] = model_tuner(BaggingClassifier(estimator=glm_en_select),None,param_grids["params_bagging"])
```

```python
models["svm_bagging"] = model_tuner(BaggingClassifier(estimator=svm_PCA),None,param_grids["params_bagging"])
```

```python
models["knn_bagging"] = model_tuner(BaggingClassifier(estimator=knn_select),None,param_grids["params_bagging"])
```

```python
models["tree_bagging"] = model_tuner(BaggingClassifier(estimator=tree_select),None,param_grids["params_bagging"])
```

----

### 2. Random Forest

```python
models["rf"] = model_tuner(RandomForestClassifier(),preprocessor_rf(),param_grids["params_rf"])
```

----

### 3. Extra Trees

```python
models["et"] = model_tuner(ExtraTreesClassifier(),preprocessor_et(),param_grids["params_et"])
```

----

### 4. Gradient Boosting

```python
models["gbc"] = model_tuner(GradientBoostingClassifier(),preprocessor_gbc(),param_grids["params_gbc"])
```

----

### 5. XGBoost

```python
models["xgb"] = model_tuner(XGBClassifier(objective="binary:logistic",
                                          eval_metric="logloss",
                                          tree_method="hist"),preprocessor_xgb(), param_grids["params_xgb"])
```

----

### 6. LightGBM

```python
models["lgbm"] = model_tuner(LGBMClassifier(objective="binary"),preprocessor_lgbm(),param_grids["params_lgbm"])
```

----

### Zapis

```python
joblib.dump(models, "all_ensamble_grids.pkl")
```
