```python
import joblib

import pandas as pd
```

----

```python
basic_models = joblib.load("all_basic_grids.pkl")

b_m_res = []

for name, (_, score, var) in basic_models.items():
    b_m_res.append([name, score, var])

b_m_res = pd.DataFrame(b_m_res, columns=["name", "score", "var"]).sort_values(by = "score", ascending=False).reset_index(drop=True)

b_m_res
```

```python
ensamble_models = joblib.load("all_ensamble_grids.pkl")

e_res = []

for name, (_, score, var) in ensamble_models.items():
    e_res.append([name, score, var])

e_res = pd.DataFrame(e_res, columns=["name", "score", "var"]).sort_values(by = "score", ascending=False).reset_index(drop=True)

e_res
```
