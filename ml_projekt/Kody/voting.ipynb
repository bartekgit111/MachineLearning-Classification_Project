```python
import joblib
from config import kfold
import numpy as np
import pandas as pd

from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score

ensamble_models = joblib.load("all_ensamble_grids.pkl")

v_estimators=[("et", ensamble_models["et"][0]),
            ("rf", ensamble_models["rf"][0]),
            ("lgbm", ensamble_models["lgbm"][0]),
            ("xgb", ensamble_models["xgb"][0])]


X_train = pd.read_csv("data/artifical_train_data.csv")
y_train = np.ravel(pd.read_csv("data/artifical_train_labels.csv")) - 1
```

```python
weights = [(ensamble_models["et"][1]),
            (ensamble_models["rf"][1]),
            (ensamble_models["lgbm"][1]),
            (ensamble_models["xgb"][1])] 

weights = weights / sum(weights)
```

```python
ranks = [4,3,2,2]
```

----
## Voting
----

### 1. Soft

```python
v = cross_val_score(
    VotingClassifier(estimators=v_estimators, voting="soft"),
    X_train,
    y_train,
    cv=kfold,
    scoring="balanced_accuracy",
    n_jobs=-1
)

print("Voting score:", np.round(v.mean(),6))
print("Voting std:", np.round(v.std(),6))
```

### 2. Soft + weights

```python
v_weights = cross_val_score(
    VotingClassifier(estimators=v_estimators, voting="soft", weights=weights),
    X_train,
    y_train,
    cv=kfold,
    scoring="balanced_accuracy",
    n_jobs=-1
)

print("Voting score:", np.round(v_weights.mean(),6))
print("Voting std:", np.round(v_weights.std(),6))
```

### 3. Soft + ranks

```python
v_ranks = cross_val_score(
    VotingClassifier(estimators=v_estimators, voting="soft", weights=ranks),
    X_train,
    y_train,
    cv=kfold,
    scoring="balanced_accuracy",
    n_jobs=-1
)

print("Voting score:", np.round(v_ranks.mean(),6))
print("Voting std:", np.round(v_ranks.std(),6))
```
